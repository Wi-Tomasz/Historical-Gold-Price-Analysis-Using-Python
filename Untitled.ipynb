### Library import

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
import yfinance as yf
from statsmodels.tsa.arima.model import ARIMA
from pmdarima import auto_arima
```

### Loading data

```python
df = pd.read_csv('Historical data gold.csv', parse_dates=['Date'])
df.set_index('Date', inplace=True)
print(df.head())
print(df.describe())
```

### Data checking and cleaning

```python
df.dtypes
```

```python
df['Price'] = df['Price'].str.replace(' ', '', regex=False)
df['Price'] = df['Price'].astype(float)
```

```python
df.dtypes
```

```python
df.tail()
```

```python
df.describe()
```

### Data visualization

```python
sns.set(style="darkgrid")
plt.figure(figsize=(14, 6))
plt.plot(df.index, df['Price'], label='Gold price (USD/oz)', color='gold')
plt.title('Gold price over time')
plt.xlabel('Date')
plt.ylabel('Price [USD]')
plt.legend()
plt.show()
```

### Statistical analysis

```python
# Daily fluctuations
df['Daily_Return'] = df['Price'].pct_change()
plt.figure(figsize=(10, 4))
sns.histplot(df['Daily_Return'].dropna(), bins=50, kde=True, color='orange')
plt.title('Distribution of daily rates of return')
plt.xlabel('Rate of return')
plt.show()
```

```python
# Statistics
print("Average daily rate of return:", df['Daily_Return'].mean())
print("Standard deviation:", df['Daily_Return'].std())
print("Cumulative return:", (1 + df['Daily_Return']).prod() - 1)
```

### Econometric analyses

```python
decomposition = seasonal_decompose(df['Price'], model='multiplicative', period=365)
decomposition.plot()
plt.show()
```

### Correlation analysis

```python
df.rename(columns={'Price': 'Gold'}, inplace=True)
tickers = {
    'SP500': '^GSPC',
    'USD_Index': 'DX-Y.NYB',
    'Oil': 'CL=F'
}

start = df.index.min().strftime('%Y-%m-%d')
end = df.index.max().strftime('%Y-%m-%d')
```

```python
df_other = yf.download(list(tickers.values()), start=start, end=end)['Close']
df_other.columns = tickers.keys() 
```

```python
df_all = pd.concat([df['Gold'], df_other], axis=1)
```

```python
df_returns = df_all.pct_change().dropna()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df_returns.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Daily Return Correlation Matrix")
plt.show()
```

### ARIMA modeling

```python
df = df.asfreq('D') 
df['Gold'] = df['Gold'].interpolate()
```

```python
df['Gold'].plot(figsize=(12, 5), title='Gold price (USD)', ylabel='USD/oz')
plt.show()
```

```python
model_auto = auto_arima(df['Gold'], seasonal=False, trace=True,
                        error_action='ignore', suppress_warnings=True)

print(model_auto.summary())
```

```python
model = ARIMA(df['Gold'], order=model_auto.order)
model_fit = model.fit()
```

```python
forecast = model_fit.forecast(steps=30)
```

```python
plt.figure(figsize=(12, 5))
plt.plot(df['Gold'], label='Historical')
plt.plot(forecast.index, forecast, label='Forecast', color='red')
plt.title('Gold Price Forecast (ARIMA)')
plt.xlabel('Date')
plt.ylabel('Price USD')
plt.legend()
plt.show()
```
